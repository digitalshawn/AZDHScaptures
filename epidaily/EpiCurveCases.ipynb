```python
import requests
import re
```

```python
data = {
    'worksheetPortSize': '{"w":1536,"h":1250}',
    'dashboardPortSize': '{"w":1536,"h":1250}',
    'clientDimension': '{"w":1536,"h":349}',
    'renderMapsClientSide': 'true',
    'isBrowserRendering': 'true',
    'browserRenderingThreshold': '100',
    'formatDataValueLocally': 'false',
    'clientNum': '',
    'navType': 'Reload',
    'navSrc': 'Top',
    'devicePixelRatio': '2.5',
    'clientRenderPixelLimit': '25000000',
    'allowAutogenWorksheetPhoneLayouts': 'true',
    'sheet_id': 'Epi%20Curve%20Dashboard',
    'showParams': '{"checkpoint":false,"refresh":false,"refreshUnmodified":false}',
    'filterTileSize': '200',
    'locale': 'en_US',
    'language': 'en',
    'verboseMode': 'false',
    ':session_feature_flags': '{}',
    'keychain_version': '1'
}
```

```python
def main2(url):
    with requests.Session() as req:
        r = req.post(url)
        sid = r.headers.get("X-Session-Id")

        r = req.post(
            f"https://www.azdhs.gov/preparedness/epidemiology-disease-control/infectious-disease-epidemiology/covid-19/dashboards/clear/sessions/{sid}")

        r = req.get(
            "https://tableau.azdhs.gov/views/EpiCurve/EpiCurveDashboard?:embed=y&:toolbar=no&%1Btabs=n")

        match = re.search(r"lastUpdatedAt.+?(\d+),", r.text).group(1)


        time = '{"featureFlags":"{\"MetricsAuthoringBeta\":false}","isAuthoring":false,"isOfflineMode":false,"lastUpdatedAt":xxx,"workbookId":9}'.replace(
            'xxx', f"{match}")

        data['stickySessionKey'] = time
        nid = r.headers.get("X-Session-Id")
        print(nid)
        r = req.post(
            f"https://tableau.azdhs.gov/vizql/w/EpiCurve/v/EpiCurveDashboard/bootstrapSession/sessions/{nid}", data=data)
        print(f"https://tableau.azdhs.gov/vizql/w/EpiCurve/v/EpiCurveDashboard/bootstrapSession/sessions/{nid}")
        
        print(r.text)
        return r


r=main2("https://tableau.azdhs.gov/views/EpiCurve/EpiCurveDashboard?:embed=y&:toolbar=no&%1Btabs=n")
```

```python
def Convert(string): 
    li = list(string.split(",")) 
    return li 
sp=r.text.find("\"dataColumns\":[{\"dataType\": \"integer\",\"dataValues\":")+52
EPI=Convert(r.text[sp:].split(']')[0])
```

```python
def Convert2(string): 
    li = list(string.split('\",\"'))
    return li
sp=r.text.find("\"dataType\": \"datetime\",\"dataValues\":")+38
dates_list=Convert2(r.text[sp:].split('\"]')[0]) 
len(dates_list)
```

```python

first=r.text.find('aliasIndices\":[],\"formatstrIndices\":[]},{\"tupleIds\":[],\"valueIndices\":')
second=r.text.find('aliasIndices\":[],\"formatstrIndices\":[]},{\"tupleIds\":[],\"valueIndices\":',first+1)+71
alias=Convert(r.text[second:].split(']')[0])
alias
```

```python
alias = [int(i) for i in alias]
EPI = [int(i) for i in EPI]
```

```python
EPI2=alias
for i in range(0,len(alias)):
    EPI2[i]= EPI[alias[i]]
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
epi_curve_cases = np.array(EPI2)\
```

```python
from datetime import date
#>>> date.fromisoformat('2019-12-04')
i=-1
for s in dates_list:
    i=i+1
    s2 = date.fromisoformat(s[0:10])
    dates_list[i]=s2
dates=np.array(dates_list)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.plot(dates,epi_curve_cases,label='cases_by_date_collected') 
plt.grid(True)
plt.title('Cases Epi-Curve')
ax.legend()
fig.autofmt_xdate()
```
